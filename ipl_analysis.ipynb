```python
total_score = ((df2.groupby(['match_id','inning']).sum()['total_runs'])+1).reset_index()
total_score
```

```python
## Preprocessing
```

```python
tm = ['Sunrisers Hyderabad', 'Mumbai Indians','Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals','Delhi Capitals']
```

```python
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
match_df = match_df[match_df['team1'].isin(tm)]
match_df = match_df[match_df['team2'].isin(tm)]
```

```python
match_df.head()
```

```python
match_df.shape
```

```python
match_df['dl_applied'].value_counts()
```

```python
match_df = match_df[match_df['dl_applied']!=1]
```

```python
match_df.head()
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
deli_df = match_df.merge(df2,on='match_id')
```

```python
deli_df = deli_df[deli_df['inning']==2]
```

```python
deli_df.head()
```

```python
deli_df.shape
```

```python
deli_df['current_run'] = deli_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
deli_df.head()
```

```python
deli_df['run_left'] = deli_df['total_runs_x'] - deli_df['current_run']
```

```python
deli_df.head()
```

```python
deli_df['ball_left'] = 126-(deli_df['over']*6 + deli_df['ball'])
```

```python
deli_df.head()
```

```python
deli_df.tail()
```

```python
deli_df['player_dismissed'] = deli_df['player_dismissed'].fillna('0')
deli_df['player_dismissed'] = deli_df['player_dismissed'].apply(lambda x:x if x=='0' else '1')
deli_df['player_dismissed'] = deli_df['player_dismissed'].astype('int')
wickets = deli_df.groupby('match_id')['player_dismissed'].cumsum().values
deli_df['wicket'] = 10 - wickets
```

```python
deli_df
```

```python
deli_df.columns
```

```python
deli_df['crr'] = (deli_df['current_run']*6)/(120 - deli_df['ball_left'])
```

```python
deli_df.head()
```

```python
deli_df['rrr'] = (deli_df['run_left']*6)/deli_df['ball_left']
```

```python
deli_df.head()
```

```python
def cal(x):
    return 1 if x['batting_team'] == x['winner'] else 0
```

```python
deli_df['result'] = deli_df.apply(cal,axis=1)
```

```python
deli_df.head()
```

```python
df = deli_df[['batting_team','bowling_team','city','run_left',
              'ball_left','wicket','total_runs_x','crr','rrr','result']]
```

```python
df = df.sample(df.shape[0])
```

```python
df.sample()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df = df[df['ball_left']!=0]
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.8,random_state=1)
```

```python
X_train.describe()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
lg = LogisticRegression(solver='liblinear')
```

```python
pipe = Pipeline(steps=[('setp1',trf),('step2',lg)])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
